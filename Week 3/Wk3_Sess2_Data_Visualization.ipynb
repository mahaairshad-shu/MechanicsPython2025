#Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It has alot of resemblance with the Matlab graphics commands.


Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

###Line Plot

A simple 2D plot using NumPy arrays

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

Let's plot data from Pandas DF

```python
import pandas as pd

data = {'Year': [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
        'Men': [9.74, 9.69, 9.58, 9.78, 9.82, 9.63, 9.77, 9.77, 9.74, 9.81, 9.82, 9.79, 9.76, 9.84, 9.80, 9.76, 9.77, 9.79]}
df = pd.DataFrame(data)

# --- Basic Line Plot ---
plt.figure(figsize=(12, 6))            # set a fig size (optional)
plt.plot(df['Year'], df['Men'])     # Plot year vs sales
plt.xlabel('Year')                   # Set label for x-axis (optional)
plt.xticks(np.arange(2007,2025))     # x axis tick marks (optional)
plt.ylabel('Time')                   # Set label for y-axis (optional)
plt.title('Male 100m best time each year')             # Set title
plt.grid(True)                        # Adds gridlines
plt.show()
```

Customization options

Detailed options can be found here: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# --- Customized Line Plot ---
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Men'], marker='o', linestyle='--', color='red', linewidth=2, label='100m') # Customize the line's appearance

plt.xlabel('Year', fontsize=14)               # Adjust label font sizes
plt.xticks(np.arange(2007,2025))               # x axis tick marks (optional)
plt.ylabel('Time', fontsize=14)
plt.title('Male 100m best time each year', fontsize=16)  # Adjust title font size

plt.legend()                                  # show legend
plt.grid(True, linestyle='--', alpha=0.7)     # Customize gridlines
plt.ylim(9.5,10)                              # set y limit

plt.annotate('Best Time', xy=(2009, 9.58), xytext=(2007, 9.8),
             arrowprops=dict(facecolor='black', shrink=0.05)) # Add an annotation


plt.show()
```

Plotting multiple lines in the same graph/figure

```python
# --- Multiple lines on the same plot---
WomensTime = [10.89, 10.78, 10.64, 10.78, 10.70, 10.70,10.71, 10.80, 10.74, 10.70, 10.85, 10.85, 10.87, 10.54, 10.62, 10.65, 10.72]

df['Women'] = pd.DataFrame(WomensTime)

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Men'], label="Men", marker="o", linestyle="-")
plt.plot(df['Year'], df["Women"], label="Women", marker="s", linestyle="--")
plt.xlabel("Year")
plt.xticks(np.arange(2007,2025))               # x axis tick marks (optional)
plt.ylabel("Time")
plt.title("Best 100m time each year")
plt.legend()
plt.show()
```

**Some other types of plots**

###Bar plot

```python
# Sample Data
data = {'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'B'],
        'Value': [10, 18, 16, 8, 9, 14, 4]}
df = pd.DataFrame(data)
df
```

```python
# --- Basic Bar Plot ---
plt.figure(figsize=(8, 6))  # Adjust figure size
plt.bar(df['Category'], df['Value'], color='skyblue')
plt.title('Basic Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# --- Customized Bar Plot ---
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Value'], color=['red', 'green', 'blue', 'purple', 'orange', 'yellow'], width=0.6)
plt.title('Customized Bar Plot', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.ylim(0, 30)  # Set y-axis limits
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid
plt.show()
```

```python
# --- Grouped Bar Plot (example)---
# Sample Data (replace with your actual grouped data)
import seaborn as sns #importing another useful visualizations library

data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
        'Value': [12, 18, 15, 22, 10, 16]}
df_grouped = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', hue='Subcategory', data=df_grouped, palette='pastel') # Use Seaborn for grouped bar plot
plt.title('Grouped Bar Plot')
plt.show()
```

###Scatter Plot

```python
# --- Scatter Plot ---
dataScatter = {'xValue': [13, 19, 14, 21, 11, 15],
        'yValue': [12, 18, 15, 22, 10, 16]}
dfScatter = pd.DataFrame(dataScatter)

plt.figure(figsize=(8, 6))
plt.scatter(dfScatter['xValue'], dfScatter['yValue'], color='orange', marker='o', s=100) #s is markersize
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.plot(np.arange(10,22,0.01), np.poly1d(np.polyfit(dfScatter['xValue'], dfScatter['yValue'], 1))(np.arange(10,22,0.01)))
plt.show()
```

###Histogram

```python
# --- Histogram ---
print(df['Category'])

plt.figure(figsize=(8, 6))
plt.hist(df['Category'], bins=5, color='purple', edgecolor='black') #bins are number of columns
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

###Box Plot

```python
# --- Box Plot ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='pastel', hue='Category' )
plt.title('Box Plot')
plt.show()
```

###Pie Chart

```python
# --- Pie Chart ---

plt.figure(figsize=(6, 6))

# Calculate the percentage of each category
category_counts = df['Category'].value_counts()
category_percentages = (category_counts / category_counts.sum()) * 100

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size if needed
plt.pie(category_percentages, labels=category_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Categories')
plt.show()
```
