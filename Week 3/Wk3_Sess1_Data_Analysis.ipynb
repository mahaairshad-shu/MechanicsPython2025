## Creating a new dataframe

Creating a DataFrame from a dictionary

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 28],
        'City': ['New York', 'London', 'Paris', 'Tokyo']}
df = pd.DataFrame(data)
df
```

Creating a DataFrame from a list of lists

```python
data = [['Alice', 25, 'New York'],
        ['Bob', 30, 'London'],
        ['Charlie', 22, 'Paris'],
        ['David', 28, 'Tokyo']]
df2 = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df2
```

 Creating a DataFrame from a list of dictionaries

```python
data = [{'Name': 'Alice', 'Age': 25, 'City': 'New York'},
        {'Name': 'Bob', 'Age': 30, 'City': 'London'},
        {'Name': 'Charlie', 'Age': 22, 'City': 'Paris'},
        {'Name': 'David', 'Age': 28, 'City': 'Tokyo'}]
df3 = pd.DataFrame(data)
df3
```

Creating a DataFrame from a NumPy array

```python
import numpy as np
data = np.array([['Alice', 25, 'New York'],
                 ['Bob', 30, 'London'],
                 ['Charlie', 22, 'Paris'],
                 ['David', 28, 'Tokyo']])

df4 = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df4
```

####Accessing columns

```python
# Accessing columns
print("\nAccessing the 'First Name' column:")
df['First Name'].head()
```

####Accessing rows using index

```python
# Accessing rows using index
print("\nAccessing the first row:")
df.iloc[0]
```

####Accessing a specific cell value

```python
# Accessing a specific cell value
print("\nAccessing the value in the first row and 'Last Name' column:")
df.loc[0, 'Last Name']
```

####Filtering data

```python
# Filtering data
print("\nFiltering for female users:")
df[df['Sex'] == 'Female'].head()
```

####Boolean indexing (more complex filtering)

```python
print("Older male users:\n")
df[(df['Sex'] == 'Male') & (df['Year of Birth'] < 1970)].head()
```

####Accessing multiple columns

```python
# Accessing multiple columns
print("\nAccessing the 'First Name' and 'Last Name' columns:")
df[['First Name', 'Last Name']].head()
```

####Accessing a slice of rows

```python
# Accessing a slice of rows
print("\nAccessing rows 1 through 3:")
df[1:4] # Note: this is exclusive of the upper limit
```

####Iterating through rows

```python
# Iterating through rows
print("\nIterating through rows:")
for index, row in df.iterrows():
    print(f"User: {row['First Name']}, Job Title: {row['Job Title']}")
```

##Manipulating data

####Columns

####Rename a column

```python
df = df.rename(columns={"Year of Birth": "YOB"})
df.columns
```

####Add a new column

```python
import numpy as np
df['Salary'] = np.random.randint(4000, 8000, size=len(df)) #Adding a Salary column
df.head()
```

####Delete a column

```python
#We can delete the Index column here as its unneccessary
del df['Index']
df.head()
```

###Rows

####Adding a new row

```python
# Adding a new row
new_row = pd.DataFrame({'First Name': ['Alice'], 'Last Name': ['Smith'], 'Sex': ['Female'], 'YOB': [1985], 'Job Title': ['Data Scientist'], 'Salary':6000})
df = pd.concat([df, new_row], ignore_index=True)
df.tail() #Check the last few rows to see if the new row was added
```

####Deleting a row

```python
# Deleting a row
df = df.drop(index=df.index[-1]) #Delete the last row
df.tail()
```

####Modifying an existing row

```python
#Modifying an existing row
df.loc[df.index[-1], 'Salary'] = 8000 #Modify the salary of the last row
df.tail()
```

####Conditional modification of rows

```python
#Conditional modification of rows
df.loc[df['Sex'] == 'Female', 'Salary'] = df.loc[df['Sex'] == 'Female', 'Salary'] * 1.1 #Increase salary of female users by 10%
df.head()
```

####Drop the rows with missing values

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
print('\ncurrent len: ', len(df))
```

```python
# drop rows with missing values
df = df.dropna() #drops all rows having a NaN value
print(df.isnull().sum())
print('\new len: ', len(df))
```

####Turn a column into Index. Instead of using numbers as row indexes.

```python
# Set a new index
df = df.set_index('First Name') # Setting the index to the "First Name" column
print("\nDataFrame with 'First Name' as index:")
df.head()
```

####The First Names can now be used to access the row

```python
# Using first names to access the row
print("\nAccessing Bridget's Last Name:")
df.loc['Bridget','Last Name']
```

####Indexes can also be sorted to sort the order of the rows

```python
#Sort index
df = df.sort_index()
print("\nDataFrame with sorted index:")
df.head()
```

###Data type

```python
# Data type checking and conversion (if needed)
print("\nData Types:")
print(df.dtypes) #strings are loaded as object type from a csv by default
```

####Convert a data type

```python
df['Salary'] = df['Salary'].astype(float) #convert to float
print(df.dtypes)
```

###Save a df as csv

```python
df.to_csv('users_dataset.csv', index=False)
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Specific statistics
print("\nMean YOB:", df['YOB'].mean())
print("Median Salary:", df['Salary'].median())
```
