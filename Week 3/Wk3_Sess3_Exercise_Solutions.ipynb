#120 years of Olympic history: athletes and results

This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016.

Note that the Winter and Summer Games were held in the same year up until 1992. After that, they staggered them such that Winter Games occur on a four year cycle starting with 1994, then Summer in 1996, then Winter in 1998, and so on. A common mistake people make when analyzing this data is to assume that the Summer and Winter Games have always been staggered.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drive_link = 'https://drive.google.com/file/d/18flGuhE2rfisV6PlA5826p189jVbPr6b/view?usp=drive_link'
url='https://drive.google.com/uc?id=' + drive_link.split('/')[-2]
df = pd.read_csv(url)

df.head()
```

```python
df.info(verbose=True)
```

###Exercise: 1
Plot number of Events, Nations and Athletes over time




```python
df2 = pd.DataFrame()
df2['Year'] = sorted(df['Year'].unique())
df2

#list comprehension using lambda to fill one column with another

#Events
df2['Events'] = df2['Year'].apply(lambda x: len(df.loc[df['Year'] == x, 'Event'].unique()))

#Nations
df2['Nations'] = df2['Year'].apply(lambda x: len(df.loc[df['Year'] == x, 'Team'].unique()))

#Athletes
df2['Athletes'] = df2['Year'].apply(lambda x: len(df.loc[df['Year'] == x, 'Name'].unique()))

df2
```

```python
# Exercise 1
# Write your code here


# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 12))  # 3 rows, 1 column

# Plot data on each subplot
axes[0].plot(df2['Year'], df2['Events'], label="Events", color='blue')
axes[0].set_title('Subplot 1')
axes[0].legend()


axes[1].plot(df2['Year'], df2['Nations'], label="Nations", color='red') # Example of customized line
axes[1].set_title('Subplot 2')
axes[1].legend()

axes[2].plot(df2['Year'], df2['Athletes'], label="Athletes", color='green') # Example of a scatter plot
axes[2].set_title('Subplot 3')
axes[2].legend()

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

###Exercise: 2
Plot number of male and female Olympians over time


```python
# Exercise 2
# Write your code here
df3 = pd.DataFrame()
df3['Year'] = sorted(df['Year'].unique())


df3['M'] = df3['Year'].apply(lambda x: 0 if 'M' not in df.loc[df['Year'] == x, 'Sex'].value_counts() else df.loc[df['Year'] == x, 'Sex'].value_counts()['M'])
df3['F'] = df3['Year'].apply(lambda x: 0 if 'F' not in df.loc[df['Year'] == x, 'Sex'].value_counts() else df.loc[df['Year'] == x, 'Sex'].value_counts()['F'])
df3
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df3['Year'], df3['M'], label="Male Count", marker="o",color = 'blue')
plt.plot(df3['Year'], df3['F'], label="Female Count", marker="o",color = 'magenta')
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Number of Male and Female athletes over time")
plt.legend()
plt.show()
```

##Exercise: 3
Based on the highest number of medals - display the gold, bronze and silver medals count for top 20 Nations

```python
# Exercise 3
# Write your code here
df4 = pd.DataFrame()
df4['Nations'] = sorted(df['Team'].unique())

df4
```

```python
df4['Gold'] = df4['Nations'].apply(lambda x: 0 if 'Gold' not in df.loc[df['Team'] == x, 'Medal'].value_counts() else df.loc[df['Team'] == x, 'Medal'].value_counts()['Gold'])
df4['Silver'] = df4['Nations'].apply(lambda x: 0 if 'Silver' not in df.loc[df['Team'] == x, 'Medal'].value_counts() else df.loc[df['Team'] == x, 'Medal'].value_counts()['Silver'])
df4['Bronze'] = df4['Nations'].apply(lambda x: 0 if 'Bronze' not in df.loc[df['Team'] == x, 'Medal'].value_counts() else df.loc[df['Team'] == x, 'Medal'].value_counts()['Bronze'])
```

```python
df4['Total'] = df4['Gold'] + df4['Silver'] + df4['Bronze']
df4_sorted = df4.sort_values(by='Total', ascending=False)
df4_sorted
```

```python
df4_top1000 = df4_sorted.head(20)  # Consider the top 1000 nations

# Create the bar plot
fig, ax = plt.subplots(figsize=(16, 6))  # Adjust figure size for better visibility

width = 0.2  # Width of the bars

x = range(len(df4_top1000['Nations']))

ax.bar(x, df4_top1000['Gold'], width, label='Gold', color='gold')
ax.bar([i + width for i in x], df4_top1000['Silver'], width, label='Silver', color = 'silver')
ax.bar([i + 2 * width for i in x], df4_top1000['Bronze'], width, label='Bronze', color = 'brown')

# Customize the plot
ax.set_xticks([i + width for i in x])
ax.set_xticklabels(df4_top1000['Nations'], rotation=90)  # Rotate x-axis labels for readability
ax.set_xlabel('Nations')
ax.set_ylabel('Medal Count')
ax.set_title('Medal Count for Top 1000 Nations')
ax.legend()
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
